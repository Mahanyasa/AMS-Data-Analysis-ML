### Attendance Management System - Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('TD Attendance - Sheet1.csv')
data.head()
```

```python
data.shape
```

```python
print("\nNumber of participants/attendees : ",data['Name'].nunique())
print("\nNames of the participants : ")
print(data['Name'].unique())
```

```python
print("Number of domains : ",data['Domain'].nunique())
data['Domain'].unique()
```

```python
dataset = data['Domain'].value_counts()
dataset.to_csv('Domains.csv')
```

```python
domain = pd.read_csv('Domains.csv')
domain.head()
```

```python
number = list(data['Domain'].value_counts())
```

```python
number
```

```python
data.groupby('Domain').size().plot(kind='pie',subplots=True)
plt.title('Strength of attendees')
```

```python
x_axis = domain['Domain']
y_axis = domain['Count']
plt.title('Domain vs Count(Participants)')
plt.xlabel('Domain')
plt.ylabel('Count')
plt.bar(x_axis,y_axis)
plt.xticks(rotation=30)
plt.show()
```
